```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data=pd.read_csv('dataset/wiki_movie.csv')
```

```python
data.head()
```

Data Cleaning


```python
data=data.drop(["Origin/Ethnicity","Director","Cast","Wiki Page"],axis=1)
```

```python
import string
def drop_special_caracters(text):
    drop=''.join(char for char in text if char not in string.punctuation)
    return drop


data['Plot']=data['Plot'].apply(drop_special_caracters)
```

```python
data['Genre']=data['Genre'].replace("unknown",'')
```

```python
data
```

```python
df1 = pd.DataFrame({'Genre': data["Genre"], 'Plot': data["Plot"] , 'Date': data["Release Year"]})
evaluate_features=df1.apply(lambda x: ' '.join(x.astype('str')), axis=1)
```

```python
data["evaluate_features"]=evaluate_features
```

```python
data
```

```python
features=data["evaluate_features"].tolist()
features
```

Install sentence Transformer

```python
!pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Load pre-trained SentenceBERT model
model = SentenceTransformer('all-mpnet-base-v2')
#Get Embeddings
query_embedding = model.encode(features)
```

```python
#similarity between movie features
cosine_scores = cosine_similarity(query_embedding)
print(cosine_scores)
```

```python
des=input("Enter the description of the movie for which you want recommendations: ")
year=input("Enter the year of release of the movie for which you want recommendations: ")
genre=input("Enter the genre of the movie for which you want recommendations: ")
k=int(input("Enter number of recommendations you want: "))



recommendations = sorted(list(enumerate(cosine_scores[data[data["Plot"] == des].index[0]])), key = lambda x:x[1], reverse = True)
print("The top "+str(k)+" recommendations for are: ", sep = "\n")
for i in range(0,k):
    print(data["Title"].loc[recommendations[i][0]])
    print("score:" +recommendations[i][1])
    print("\n")
```
